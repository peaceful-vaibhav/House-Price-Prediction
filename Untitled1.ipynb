```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data
```

```python
# dropping id, price, date because the model will predict the price itself and the id and the date are not so much significant

train = data.drop(["id", "price", "date", "yr_renovated"], axis = 1)
test = data["price"]
```

```python
x_train, y_train, x_test, y_test = train_test_split(train, test, test_size = 0.2, random_state = 2)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train, x_test)
```

```python
prediction = regressor.predict(y_train)
```

```python
prediction
```

```python
regressor.score(y_train, y_test)
```

```python
# Using Gradient Boosting for increasing the score of the model
```

```python
from sklearn import ensemble
```

```python
reg = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, learning_rate = 0.1, loss = 'ls', min_samples_split = 2)
```

```python
reg.fit(x_train, x_test)
```

```python
reg.score(y_train, y_test)
```
